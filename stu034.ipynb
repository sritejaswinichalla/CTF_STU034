```python
import pandas as pd
import numpy as np
import hashlib
```

```python
books = pd.read_csv("books.csv")
reviews = pd.read_csv("reviews.csv")
```

```python
seed = "STU034"        
full_hash = hashlib.sha256(seed.encode()).hexdigest()
my_hash8 = full_hash[:8].upper()
print("MY_HASH (8 chars, uppercase):", my_hash8)
print("FLAG2 format =>", f"FLAG2{{{my_hash8}}}")
```

```python
books.info()
```

```python
reviews.info()
```

```python
found_reviews_text = reviews[
    reviews["text"].str.contains(my_hash8, case=False, na=False)
]

found_reviews_title = reviews[
    reviews["title"].str.contains(my_hash8, case=False, na=False)
]
```

```python
print("Matches in text:", len(found_reviews_text))
print("Matches in title:", len(found_reviews_title))
```

```python
print("Number of matching fake reviews found:", len(found_reviews_text))
found_reviews_text
```

```python
# Extract the matching review's parent_asin
target_parent_asin = found_reviews_text["asin"].iloc[0]
print("Matched Parent ASIN:", target_parent_asin)
```

```python
books.columns
```

```python
my_book = books[books["parent_asin"] == target_parent_asin]

print("Matched Book Row:")
my_book[["title", "rating_number", "average_rating", "parent_asin"]]
```

```python
title = my_book["title"].iloc[0]
print("Book Title:", title)

clean = "".join(title.split())

first8 = clean[:8]
print("First 8 non-space chars:", first8)

FLAG1 = hashlib.sha256(first8.encode()).hexdigest()
print("FLAG1 =", FLAG1)
```
